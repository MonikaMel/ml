```python
import numpy as np
import pandas as pd
```

```python
from sklearn import preprocessing
```

```python
unnormalized_data = pd.read_csv("data.csv")
unnormalized_data = unnormalized_data.sample(frac=1).reset_index(drop=True)
scaler = preprocessing.MinMaxScaler()
col_names = unnormalized_data.columns
df = scaler.fit_transform(unnormalized_data)
data = pd.DataFrame(df, columns=col_names)
data
```

```python
train_data = data.iloc[:1251]
train_data
```

```python
test_data = data.iloc[1251:]
test_data
```

```python
train_data.insert(0, "std", [1]*1251)
train_data
```

```python
test_data.insert(0, "std", [1]*(len(data)-1251))
test_data
```

```python
alpha = 1e-3
```

```python
def hypotesis(thetas , i, X):
    thetaTx = 0
    for j in range(len(thetas)):
        thetaTx += thetas[j] * X.iloc[i, j]
    res = 1/(1 + np.exp(-thetaTx))
    return res

def J(j, thetas, X):
    res = 0
    for i in range(41):
        res += (hypotesis(thetas, i, X) - X.iloc[i, len(thetas)])*X.iloc[i, j]
    return res

def logistic_reg(X, alpha):
    thetas = [0] * (len(X.columns)-1)
    for _ in range(300):
        temp_thetas = thetas[:]
        for j in range(len(thetas)):
            temp_thetas[j] -= alpha*J(j, thetas, X)
        thetas = temp_thetas[:]
    return thetas

thetas = logistic_reg(train_data, alpha)    
```

```python
thetas
```

```python
def prediction(X, thetas):
    l = []
    for i in range(len(X)):
        if hypotesis(thetas, i, X)>=0.5:
            l.append(1)
        else:
            l.append(0)
    return l

test = prediction(test_data, thetas)
```
